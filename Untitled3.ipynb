```python
import sqlite3

connexion = sqlite3.connect('factbook.db')

birth_rate_unique = connexion.execute("SELECT DISTINCT birth_rate FROM facts;").fetchall()
print(birth_rate_unique)

connexion.close()
```

```python
connexion = sqlite3.connect('factbook.db')

pop_SUM_1M = connexion.execute("SELECT SUM(DISTINCT population) FROM facts WHERE area_land > 1000000;").fetchall()

pop_SUM_1MClean = pop_SUM_1M[0][0]
print(pop_SUM_1MClean)

connexion.close()
```

```python
connexion = sqlite3.connect('factbook.db')

pop_growth_1MD = connexion.execute("SELECT population_growth / 1000000.0 FROM facts;").fetchall()

print(pop_growth_1MD)

connexion.close()
```

```python
connexion = sqlite3.connect('factbook.db')

pop_1y = connexion.execute("SELECT population * (population_growth + 1 ) FROM facts;").fetchall()

print(pop_1y)

connexion.close()
```
